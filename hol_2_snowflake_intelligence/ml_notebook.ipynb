# HOL 2: Predictive Analytics with Snowflake ML

## Build two ML models using Cortex Code skills

In this lab you'll invoke two Cortex Code skills that guide you through building ML models natively in Snowflake:

1. **Time-Series Forecasting** — Predict future loan application volumes
2. **Loan Default Classification** — Predict which loans will have late payments

**How to use this notebook:**
1. Run the setup cell below
2. Invoke each skill using `/loan-demand-forecast` and `/loan-default-classifier` in Cortex Code
3. The skill will generate a complete notebook workflow for you - review and run each cell

**Prerequisites:** HOL 1 Silver and Gold layer tables must exist.
